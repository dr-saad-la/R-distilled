## Installing Python:  Prerequisite for Installing Jupyter 

- Python is a requirement for installing the `Jupyter Notebook`. After installing Python, the Python package manager `pip` will be installed automatically, which is the tool we use to install  `Jupyter notebook` or `Jupyterlab`. 


- The Python version required differs between Jupyter Notebook releases (e.g. Python 3.6+ for Notebook v6.3, and Python 3.7+ for Notebook v7) . [source](https://docs.jupyter.org/en/latest/install/notebook-classic.html#prerequisite-python)


### Installing Classic Jupyter Notebook 

- First let us ensure that you have `pip` Python package manager by running:

```bash
pip --version
```

- If it does not produce an error, it means your machine is set up correctly, if not, you have to follow these steps to add Python to the path (the path is the location where the system searches for executable files):

- **Adding Python to the path**:
     - You can do this either from the command prompt as follows:
         - Open command prompt in administrator mode (press `windows key + R`), then type `cmd` in the Run dialog, then press `ctrl + Enter`
         - Then just type this command `setx /m "python-path:%PATH%"`
         - replace `python-path` with the location of Python on your system
         - Close up the command prompt and open it again
         - type `python --version`
         - If returned Python version, then Python has been added to the path correctly. 
         
     - Adding Python to the path using graphical user interface:
         - Open the run dialog (this is the easiest way)
         - Type `sysdm.cpl` in the Run dialog and press enter
         - The System properties will pop up
         - Go advanced tab 
         - Press on Environment variables
         - Go to System variable and find `path` and press on edit (This will add Python system wide)
         - Press on `new`
         - Add the location of the executable Python, this location looks like this `C:\Program Files\Python3.x\scripts`
         - Press ok 
         - You can add Python to this specific user the same way you did with system variable from `User variables`
         - When you finish, press ok, the ok from the environment variables. 
         - Test that Python has been added to the path from the command prompt `python --version`. 

- If you followed the steps above of installing Python and making sure it is recognized from the command prompt, then you are good to go, otherwise you should go up the steps illustrated above. 

- We can install the classic `Jupyter notebook` by running the following command:

```bash
pip install jupyter 
```

or 

```bash 
pip install notebook
```


- You can test that jupyter is installed by running the next command:

```bash
jupyter --version
```

- This command will show information `jupyter` version installed with extra information that it will be clear in a later stage.

## Registering the kernel in the current R installation

- After installing the `IRkernel`, we need to add to `Jupyter` which can be achieved by installing a kernel spec using `installspec()` function from the `IRkenel` package. This allows Jupyter to recognize the kernel and work with it interactively:

- As a programmer, or a future programmer, you don't want to take everything for granted. If you are new to R at all (and I suppose you are), what we are going to do here will not be clear to you, but this concerns the installation on the system more than the language.

- Here will try to understand what the `installspec()` does when calling it, and can customize our `Rkernel` or accept the defaults if you are happy with them. 

- In the next piece of code, we will load the `IRkernel` to the working environment, then call another function `args()` to display the arguments of the `installspec()` function. You can alternatively run the `help(installspec()` for more information. However, here I will explain that to you. 

- Getting the help of what you are doing: 


```r
> library("IRkernel")
> args(installspec)
function (user = NULL, name = "ir", displayname = "R", rprofile = NULL, 
    prefix = NULL, sys_prefix = NULL, verbose = getOption("verbose")) 
```

- `user = NULL`: to install the kernel for the current user only. You can set this argument to `user=FALSE` to install the kernel system wide.
- `name = "ir"`: This indicates the name of the kernel.  
- `displayname = "R"`: This the name which is displayed in the notebook (default: "R")). 

- Now it is time to make Jupyter aware of R kernel, I will accept the default except for the `displayname` which I will add The version (This makes me aware of what version I am working with, also it will be help debug some errors when they occur)

```r
installspec(displayname = "R 4.3") # R 4.3 is the latest version of writing this lesson
```

- You will see some programmers using this in one step by using what is called `namespace`, which means we call `installspec()` function directly from where it locates without the need to load the `IRkernel` package. I did it in two steps for the purpose of education, you can do it this way as well:

```r
> IRkernel::installspec(displayname = "R 4.2")
```                       

### Note

> By not setting different arguments, multiple calls will overwrite the kernel with a kernel spec pointing to the last R interpreter you called that commands from. 


### Installing `R` Versions  (optional and Advanced)

- You can install kernels for multiple versions of R by supplying a name and displayname argument to the `installspec()` call (You still need to install these packages in all interpreters you want to run as a jupyter kernel!):


```r
# in R 3.3
IRkernel::installspec(name = 'ir33', displayname = 'R 3.3')
# in R 3.2
IRkernel::installspec(name = 'ir32', displayname = 'R 3.2')
```

### Something Extra 

- You can install R from github using these commands. The first command installs the necessary libraries. 

- The `devtools` is the library used to install `IRkernel` from the its repository on github. 

- The last command will register the kernel for jupyter. 

```bash
install.packages(c('repr', 'IRdisplay', 'pbdZMQ', 'devtools'))
devtools::install_github('IRkernel/IRkernel')
IRkernel::installspec()  # to register the kernel in the current R installation
```
